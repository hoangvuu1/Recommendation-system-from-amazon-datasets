```python
import faiss
import numpy as np
import pandas as pd
import pickle

index = faiss.read_index("faiss/products_hnsw.index")

with open("faiss/products_ids.pkl", "rb") as f:
    item_ids = pickle.load(f)

embeddings = np.load("embedding_matrix/embeddings.npy")

k = 10
D, I = index.search(embeddings, k)

rows = []

for i in range(len(item_ids)):
    for j in range(k):
        similar_idx = I[i][j]
        score = float(D[i][j])

        item_id = item_ids[i]
        similar_item_id = item_ids[similar_idx]

        if item_id == similar_item_id:
            continue

        rows.append((item_id, similar_item_id, score))

df = pd.DataFrame(rows, columns=[
    "item_id",
    "similar_item_id",
    "similarity_score"
])


min_s = df["similarity_score"].min()
max_s = df["similarity_score"].max()

if max_s != min_s:
    df["similarity_score"] = (df["similarity_score"] - min_s) / (max_s - min_s)

df.to_parquet("item_similarity.parquet", index=False)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
from pyspark.sql import SparkSession
```

```python
spark = (
    SparkSession.builder
    .appName('CollaborativeFiltering')
    .config('spark.driver.memory', '40g')
    .config('spark.sql.shuffle.partitions', '200')
    .getOrCreate()
)
spark
```

```python
ratings = pd.read_parquet("rating_rs.parquet")


ratings = ratings.rename(columns={
    "product_id": "item_id",
    "review_rating": "rating"
})

ratings = ratings[["user_id", "item_id", "rating"]]


ratings["rating"] = ratings["rating"] / ratings["rating"].max()



sim_df = pd.read_parquet("item_similarity.parquet")



df = ratings.merge(sim_df, on="item_id", how="inner")


df["score"] = df["rating"] * df["similarity_score"]

content = (
    df.groupby(["user_id", "similar_item_id"])["score"]
    .sum()
    .reset_index()
)

content.columns = ["user_id", "product_id", "content_score"]



user_items = ratings.groupby("user_id")["item_id"].apply(set).to_dict()

content = content[
    ~content.apply(
        lambda x: x["product_id"] in user_items.get(x["user_id"], set()),
        axis=1
    )
]



min_s = content["content_score"].min()
max_s = content["content_score"].max()

if max_s != min_s:
    content["content_score"] = (
        (content["content_score"] - min_s) / (max_s - min_s)
    )



content = content.sort_values(
    ["user_id", "content_score"],
    ascending=[True, False]
)

content = content.groupby("user_id").head(20)

content.to_parquet("content_recommendations.parquet", index=False)
```

```python
ratings_raw = spark.read.parquet('rating_rs.parquet')

print(f"Total ratings: {ratings_raw.count():,}")

ratings_raw.show(3, truncate=False)


als_recs = spark.read.parquet('als_recommendations.parquet')
print("\n=== ALS Recommendations ===")
print(f"Total recommendations: {als_recs.count():,}")
als_recs.show(3, truncate=False)
```

```python
item_popularity = (
    ratings_raw.groupBy("product_id")
    .agg(
        count("review_rating").alias("rating_count"),
        avg("review_rating").alias("avg_rating")
    )
    .withColumn("popularity_score", col("rating_count") * col("avg_rating"))
    .orderBy(desc("popularity_score"))
    .limit(1000)  # Top 1000
)


popular_items = (
    item_popularity
    .select("product_id", col("popularity_score").alias("score"))
    .limit(100)  # Top 100 for cold start
)

print(f"Popular items (top 100): {popular_items.count()}")

popular_items.show(10, truncate=False)


top_popular = popular_items.select("product_id")
```

```python
content_recs = spark.read.parquet('content_recommendations.parquet')


def normalize_scores(scores_df, score_column="score", group_id="user_id"):
    """
    Normalize scores to [0, 1] range per user group
    Formula: (score - min) / (max - min)
    """
    window_spec = Window.partitionBy(group_id)

    min_score = min(score_column).over(window_spec)
    max_score = max(score_column).over(window_spec)


    normalized = when(
        max_score == min_score,
        lit(1.0)
    ).otherwise(
        (col(score_column) - min_score) / (max_score - min_score)
    )

    return scores_df.withColumn(f"{score_column}_norm", normalized)


als_normalized = normalize_scores(als_recs, "als_score", "user_id")
content_normalized = normalize_scores(content_recs, "content_score", "user_id")


als_normalized.select("user_id", "product_id", "als_score", "als_score_norm").show(5, truncate=False)


content_normalized.select("user_id", "product_id", "content_score", "content_score_norm").show(5, truncate=False)
```

```python

ALS_WEIGHT = 0.7
CONTENT_WEIGHT = 0.3


hybrid = (
    als_normalized.select("user_id", "product_id", "als_score_norm")
    .join(
        content_normalized.select("user_id", "product_id", "content_score_norm"),
        on=["user_id", "product_id"],
        how="outer"
    )
)

hybrid = hybrid.withColumn(
    "hybrid_score",
    (
        ALS_WEIGHT * coalesce(col("als_score_norm"), lit(0.0)) +
        CONTENT_WEIGHT * coalesce(col("content_score_norm"), lit(0.0))
    )
).select("user_id", "product_id", "hybrid_score")

print(f"Hybrid scores computed: {hybrid.count():,}")

hybrid.select("hybrid_score").summary("count", "min", "max", "mean").show()
```

```python
K = 10

window_spec = Window.partitionBy("user_id").orderBy(desc("hybrid_score"))

final_recommendations = (
    hybrid
    .withColumn("rank", row_number().over(window_spec))
    .filter(col("rank") <= K)
    .select(
        "user_id", "product_id",
        col("hybrid_score").alias("final_score")
    )
)

print(f"Final recommendations (top-{K} per user): {final_recommendations.count():,}")
unique_users = final_recommendations.select('user_id').distinct().count()
print(f"Unique users: {unique_users:,}")
print(f"Avg recommendations per user: {final_recommendations.count() / unique_users:.1f}")

final_recommendations.show(15, truncate=False)
```

```python
comparison_df = (
    als_normalized
    .join(
        content_normalized.select("user_id", "product_id", "content_score_norm"),
        on=["user_id", "product_id"],
        how="inner"
    )
    .withColumn(
        "hybrid_score",
        lit(ALS_WEIGHT) * col("als_score_norm") + lit(CONTENT_WEIGHT) * col("content_score_norm")
    )
    .select(
        "user_id", "product_id",
        col("als_score_norm").alias("als_norm"),
        col("content_score_norm").alias("content_norm"),
        "hybrid_score"
    )
)

print(f"Comparison data: {comparison_df.count():,} rows")

print("\nScore Distribution Statistics")
stats = comparison_df.select(
    [min(c).alias(f"min_{c}") for c in ["als_norm", "content_norm", "hybrid_score"]] +
    [max(c).alias(f"max_{c}") for c in ["als_norm", "content_norm", "hybrid_score"]] +
    [avg(c).alias(f"avg_{c}") for c in ["als_norm", "content_norm", "hybrid_score"]]
)
stats.show(1, truncate=False)

comparison_df.orderBy(desc("hybrid_score")).limit(20).show(20, truncate=False)
```
